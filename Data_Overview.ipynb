STNL DATA READ

```python
import pandas as pd
import numpy as np

stnl1 = pd.read_csv("/Users/dan/calpoly/BusinessAnalytics/GSB503HAWKS/data/stnl_activity_subset_1.csv")
stnl2 = pd.read_csv("/Users/dan/calpoly/BusinessAnalytics/GSB503HAWKS/data/stnl_activity_subset_2.csv")
stnl3 = pd.read_csv("/Users/dan/calpoly/BusinessAnalytics/GSB503HAWKS/data/stnl_activity_subset_3.csv")
stnl4 = pd.read_csv("/Users/dan/calpoly/BusinessAnalytics/GSB503HAWKS/data/stnl_activity_subset_4.csv")

stnl = pd.concat([stnl1, stnl2, stnl3, stnl4], ignore_index=True)

#print(stnl.info())

stnl_as = stnl['ACTIVITY_SOURCE'].value_counts(normalize=True) * 100
stnl_ac = stnl['ACTIVITY_CODE'].value_counts(normalize=True) * 100
stnl_at = stnl['ACTIVITY_TYPE'].value_counts(normalize=True) * 100
stnl_en = stnl['IS_ENGAGEMENT_ACTIVITY'].value_counts(normalize=True) * 100
stnl_descibe = [stnl_as, stnl_ac, stnl_at, stnl_en]

stnl.head()
```
